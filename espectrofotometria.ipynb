# **Graficos**

Se graficaron los espectros de las muestras experimentales obtenidas, tambien se hallo los 10 picos mas altos de cada una de las muestras.

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.signal import find_peaks

def graficar_picos(archivo, num_picos,name):
    # Cargar el archivo .txt en un DataFrame
    data = pd.read_csv(archivo, sep=';', header=None, names=['Longitud de Onda', 'Reflectancia'])

    # Encontrar todos los picos en los datos de reflectancia
    peaks, _ = find_peaks(data['Reflectancia'])

    # Obtener los valores de reflectancia en los picos encontrados
    peak_values = data['Reflectancia'][peaks]

    # Seleccionar los índices de los picos más altos según el número especificado
    top_peaks_indices = peak_values.nlargest(num_picos).index

    # Imprimir los valores de los picos más altos seleccionados
    print(f"Los {num_picos} picos más altos encontrados en la gráfica:")
    for index in top_peaks_indices:
        print(f"Longitud de Onda: {data['Longitud de Onda'][index]}, Reflectancia: {data['Reflectancia'][index]}")

    # Crear la gráfica de reflectancia con los picos más altos marcados
    plt.figure(figsize=(10, 6))
    plt.plot(data['Longitud de Onda'], data['Reflectancia'], label='Reflectancia', color='blue')
    plt.plot(data['Longitud de Onda'][top_peaks_indices], data['Reflectancia'][top_peaks_indices], 'ro', label=f'Top {num_picos} Picos')  # Marcar los picos más altos con puntos rojos
    plt.xlabel('Longitud de Onda (nm)')
    plt.ylabel('Reflectancia')
    plt.title(f'Espectro de Reflectancia de la muestra de {name}')
    plt.legend()
    plt.grid()
    plt.show()

```

```python
graficar_picos('reference.txt', 10,'referencia')
```

```python
graficar_picos('dark.txt', 10,'oscura')
```

```python
graficar_picos('aluminio2.txt', 10,'aluminio')
```

```python
graficar_picos('silicio.txt', 10,'silicio')
```
